```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
## SINE WAVE
cycles = 2      
samples = 360
max_amplitude = 5

xs = np.arange(samples * cycles)
ys = [] 
for x in xs:
    ys.append((np.sin(2 * np.pi * x / samples) + 1) / 2 * max_amplitude)

plt.title("Sine wave - " + str(cycles) + " cycle(s) @ " + str(samples) + " samples / cycle and max amplitude of " + str(max_amplitude) + " volts")
plt.xlabel("Output samples")
plt.ylabel("Voltage")
plt.plot(xs, ys)
plt.show()
```

```python
## SQUARE WAVE
cycles = 1
samples = 360
max_amplitude = 5

xs = np.arange(samples * cycles)
ys = [] 
y = 0
for x in xs:
    if x >= samples / 2:
        y = max_amplitude
    else:
        y = 0
    ys.append(y)


plt.title("Square wave - " + str(cycles) + " cycle(s) @ " + str(samples) + " samples / cycle and max amplitude of " + str(max_amplitude) + " volts")
plt.xlabel("Output samples")
plt.ylabel("Voltage")
plt.plot(xs, ys)
plt.show()
```

```python
## SAWTOOTH WAVE - RAMP UP
cycles = 2          
resolution = 360
max_amplitude = 5

xs = np.arange(resolution * cycles)
y_step = max_amplitude / resolution
ys = [] 
y = 0
for x in xs:
    if y >= max_amplitude:
        y = 0 
    
    y += y_step
    ys.append(y)

plt.title("Ramp up wave - " + str(cycles) + " cycle(s) @ " + str(samples) + " samples / cycle and max amplitude of " + str(max_amplitude) + " volts")
plt.xlabel("Output samples")
plt.ylabel("Voltage")

plt.plot(xs, ys)
plt.show()
```

```python
## SAWTOOTH WAVE - RAMP DOWN
cycles = 2
resolution = 360
max_amplitude = 5

xs = np.arange(resolution * cycles)
y_step = max_amplitude / resolution
ys = [] 
y = 0
for x in xs:
    if y <= 0:
        y = max_amplitude
    
    y -= y_step
    ys.append(y)

plt.title("Ramp down wave - " + str(cycles) + " cycle(s) @ " + str(samples) + " samples / cycle and max amplitude of " + str(max_amplitude) + " volts")
plt.xlabel("Output samples")
plt.ylabel("Voltage")
plt.plot(xs, ys)
plt.show()
```

```python
## RANDOM NOISE - NORMALLY DISTRIBUTED
import random

random.seed(42)
cycles = 1
samples = 360
max_amplitude = 5
mu = max_amplitude / 2
sigma = max_amplitude / 8

xs = np.arange(samples * cycles)
ys = []
for x in xs:
    ys.append(random.gauss(mu, sigma))

plt.title("Gaussian noise - " + str(cycles) + " cycle(s) @ " + str(samples) + " samples / cycle and max amplitude of " + str(max_amplitude) + " volts")
plt.xlabel("Output samples")
plt.ylabel("Voltage")

plt.plot(xs, ys)
plt.show()
```
